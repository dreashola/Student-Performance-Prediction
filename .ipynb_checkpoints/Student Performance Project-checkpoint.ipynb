```python
cat_attributes = maths_df.select_dtypes(object).columns.to_list()

# g = sns.FacetGrid(data = maths_df, palette = cat_attributes)
# g.map(plt.scttaer, )
fig, axes = plt.subplots(4,4, figsize = (12,10))
x = 0
y = 0
# maths_df['sex'].value_counts().plot(kind = 'bar', ax = axes[1][0])
for attribute in cat_attributes[:-1]:
    try:
        maths_df[attribute].value_counts().plot(kind = 'bar', ax = axes[x,y])
        y+=1
    except IndexError:
        x+=1
        y = 0
        maths_df[attribute].value_counts().plot(kind = 'bar', ax = axes[x,y])
# #     break
# plt.tight_layout()
plt.show();
```

```python
# split the data into train and test sets

maths_train_set, maths_test_set = train_test_split( maths_df, test_size=0.2, random_state=42)
```

## Explore the Data

```python
# make a copy of the train set

maths_df = maths_train_set.copy()
```

```python
# look for correlation

corr_matrix = maths_df.corr(numeric_only = True)
corr_matrix['G3'].sort_values(ascending = False)
```

```python
attributes = ['G3', 'G2', 'G1', 'failures', 'Medu']
scatter_matrix(maths_df[attributes], figsize = (12,8))
plt.show();
```

```python
# take a closer look at G1 and G2 scatterplots

fig, axes = plt.subplots(1,2, figsize = (15,5))

maths_df.plot(kind='scatter',x = 'G1', y ='G3', alpha = 0.3, grid = True, ax = axes[0])
maths_df.plot(kind='scatter',x = 'G2', y ='G3', alpha = 0.3, grid = True, ax = axes[1])
plt.show();
```
